```python
!pip install -q python-dotenv google-generativeai
import os
import time
import google.generativeai as genai
from dotenv import load_dotenv
from google.colab import files

if not os.path.exists(".env"):
    with open(".env", "w") as f:
        f.write("GEMINI_API_KEY=API+KEY")  #Replace the API key

load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")
if not API_KEY:
    raise ValueError("API Key not found")

genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("gemini-1.5-flash")

print(" Please upload a text file")
uploaded = files.upload()
if not uploaded:
    raise ValueError("No file uploaded")

filename = list(uploaded.keys())[0]
with open(filename, "r", encoding="utf-8") as file:
    text_content = file.read()

prompt = f"Summarize the following text concisely:\n\n{text_content[:5000]}"
start_time = time.time()

response = model.generate_content(
    prompt,
    generation_config={
        "temperature": 0.7,
        "max_output_tokens": 100,
        "top_k": 50,
        "top_p": 0.9
    }
)

end_time = time.time()

print("\n**Summary:**\n")
print(response.text)
time_taken = end_time - start_time
print(f"\n **Speed of Generation:** {time_taken:.2f} seconds")

accuracy_test_prompt = "What is 2+3? Also, calculate sin(30Â°) + cos(60Â°) + 12."
accuracy_response = model.generate_content(accuracy_test_prompt)

print("\n**Accuracy Test:**")
print(accuracy_response.text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Model Name": ["Gemini Flash 1.5", "Gemini Flash 2.0", "Gemini Pro 1.5 pro latest "],
    "Speed (Seconds)": [2.66, 3.37, 4.48],
    "Input Tokens": [462, 462, 462],
    "Output Tokens": [100, 100, 100],
    "Accuracy (%)": [100, 100, 100]
}

df = pd.DataFrame(data)

plt.style.use("dark_background")

fig, ax = plt.subplots(figsize=(8, 5))
ax.axis("tight")
ax.axis("off")
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc="center", loc="center")

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
for i in range(len(df) + 1):
    for j in range(len(df.columns)):
        table[(i, j)].set_facecolor("#222222")
        table[(i, j)].get_text().set_color("white")

plt.title(" Gemini Model Benchmark Comparison", fontsize=14, color="white", pad=20)
plt.show()
```
