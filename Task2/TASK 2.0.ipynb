```python
import google.generativeai as genai
import os
import pandas as pd
import time
import zipfile
from dotenv import load_dotenv
from google.colab import files

#  .env file
load_dotenv(dotenv_path="./.env")
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
if not GEMINI_API_KEY:
    raise ValueError("GEMINI_API_KEY is missing. Please check your .env file.")

#  Gemini API
genai.configure(api_key=GEMINI_API_KEY)

#  prompt versions
PROMPTS = [
    """You are an expert in sentiment analysis. Categorize the following feedback as:
    - POSITIVE
    - NEGATIVE
    - NEUTRAL
    Feedback: "{feedback}"
    Only return one of these words.""",

    """Analyze the sentiment of the given user feedback. Reply with:
    - POSITIVE if the feedback is appreciative.
    - NEGATIVE if the feedback is critical.
    - NEUTRAL if the feedback is neither positive nor negative.
    Feedback: "{feedback}"
    Respond with only one category: POSITIVE, NEGATIVE, or NEUTRAL.""",

    """Classify the following feedback into one of these categories: POSITIVE, NEGATIVE, NEUTRAL.
    Feedback: "{feedback}"
    Provide only the sentiment category without additional explanation."""
]

def analyze_sentiment(feedback, prompt):
    """Uses Gemini AI to analyze sentiment with a given prompt."""
    formatted_prompt = prompt.format(feedback=feedback)
    try:
        model = genai.GenerativeModel("gemini-1.5-flash")
        response = model.generate_content(formatted_prompt)
        return response.text.strip()
    except Exception as e:
        return f"Error: {str(e)}"

# extract ZIP file
print("Please upload your ZIP file containing feedback CSV.")
uploaded = files.upload()
zip_file_name = list(uploaded.keys())[0]

# Extract files
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall("./extracted_data")
    extracted_files = zip_ref.namelist()

#  CSV file
csv_file_name = None
for file in extracted_files:
    if file.endswith(".csv"):
        csv_file_name = f"./extracted_data/{file}"
        break
if not csv_file_name:
    raise ValueError("No CSV file found in the uploaded ZIP.")

# Read CSV file with proper encoding
try:
    df = pd.read_csv(csv_file_name, encoding='ISO-8859-1')  # Fix for encoding error

    if 'feedback' not in df.columns:
        raise ValueError("The CSV file must contain a column named 'feedback'.")

    feedbacks = df['feedback'].dropna().tolist()

    # Iterate through different prompts
    for i, prompt in enumerate(PROMPTS):
        results = []

        print(f"\nProcessing feedback with Prompt {i+1}...\n")

        for feedback in feedbacks:
            sentiment = analyze_sentiment(feedback, prompt)
            results.append({'feedback': feedback, 'LLM response': sentiment})
            print(f"Feedback: {feedback}\nSentiment: {sentiment}\n")
            time.sleep(5)  # Prevent hitting API rate limits

        # Save results to an Excel file with two sheets
        output_file = f"sentiment_results_prompt_{i+1}.xlsx"
        with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
            results_df = pd.DataFrame(results)
            results_df.to_excel(writer, sheet_name="Sentiment Analysis", index=False)
            prompt_df = pd.DataFrame({'Prompt': [prompt]})
            prompt_df.to_excel(writer, sheet_name="Prompt Used", index=False)

        # Auto-download the file
        files.download(output_file)
        print(f"\nSentiment analysis for Prompt {i+1} completed! Results saved to '{output_file}' and downloaded.")

except Exception as e:
    print(f"Error reading CSV file: {e}")
```
